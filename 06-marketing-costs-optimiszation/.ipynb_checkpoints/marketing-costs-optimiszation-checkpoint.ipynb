#### Retention Rate

```python
first_session = visits.groupby(['uid']).agg({'start_ts':'min'})
first_session.columns=['first_session']
# вывод месяца первой сессии
```

```python
user_activity = visits.join(first_session, on='uid')
user_activity['first_session'] = user_activity['first_session'].astype('datetime64[M]')
# добавление первого месяца к таблице
```

```python
user_activity['lifetime'] = ((user_activity['session_month'] - user_activity['first_session']) / np.timedelta64(1,'M')).round().astype('int')
# вычисление времени жизни когорт
```

```python
cohorts = user_activity.groupby(['first_session', 'lifetime']).agg({'uid':'nunique'}).reset_index()
# количество уникальных пользователей каждый месяц
```

```python
inital_users_count = cohorts[cohorts['lifetime'] == 0][['first_session','uid']]
inital_users_count = inital_users_count.rename(columns={'uid':'cohort_users'})
cohorts = cohorts.merge(inital_users_count,on='first_session')
# исходное количество пользователей в когорте
```

```python
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users']
# Retention Rate
```

```python
retention_pivot = cohorts.pivot_table(index='first_session',columns='lifetime',values='retention',aggfunc='sum').drop(0, 1).dropna(how='all')
retention_pivot.index = pd.to_datetime(retention_pivot.index)
# создание сводной таблицы
```

```python
plt.figure(figsize=(16, 7))
plt.title('Когорты: Удержание пользователей')
ax = sns.heatmap(retention_pivot, annot=True, fmt='.2%', linewidths=1, cmap="YlGnBu", vmax=0.15)
plt.ylabel('Месяц первой сессии')
plt.xlabel('Прошло месяцев с первой сессии')
ax.set_yticklabels(retention_pivot.index.date);
# тепловая карта
```

Средний Retention Rate на второй месяц жизни когорт.

```python
round(retention_pivot[1].mean(), 2)
```

**Вывод**
- Retention Rate в первый месяц убывает по когортам с течением времени. За 11 месяцев коэффициент удержания упал на 3,1%. Это может говорить о привлечении нецелевой аудитории или об изменениях на сайте, которые не нравятся пользователям.
- В основном (90-95%) люди отваливаются после первого месяца

#### LTV

```python
order_cohorts = orders.merge(first_order, on='uid', how='inner')
order_cohorts['first_order_month']=order_cohorts['first_order'].astype('datetime64[M]')
# добавление месяца первого заказа
```

```python
cohort_sizes = (order_cohorts.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index())
cohort_sizes.columns = ['first_order_month', 'n_buyers']
# вывод размеров когорт
```

```python
cohorts = (order_cohorts.groupby(['first_order_month', 'month']).agg({'revenue': 'sum'}).reset_index())
# вывод доходов по месяцам
```

```python
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
# добавление размеров когорт
```

```python
report['age'] = (report['month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
# вывод возраст когорт
```

```python
report['ltv'] = report['revenue'] / report['n_buyers']
ltv = report.pivot_table(index='first_order_month', columns='age', values='ltv', aggfunc='mean').round(2)
ltv = ltv.cumsum(axis = 1)
# вывод LTV
```

```python
ltv
```

```python
plt.figure(figsize=(16, 7))
plt.title('LTV')
ax = sns.heatmap(ltv, annot=True, fmt='.2f', linewidths=1, cmap="YlGnBu")
plt.ylabel('Месяц первой покупки')
plt.xlabel('Прошло месяцев с первой покупки')
ax.set_yticklabels(ltv.index.date)
plt.show()
# тепловая карта
```

```python
ltv_6_month=ltv.iloc[[0,1,2,3,4,5,6]]
```

```python
print('Средний LTV за 6 месяцев ', round(ltv_6_month[5].mean(), 2))
```

Самой прибыльной когортой оказалась 2017-09-01 и самая первая из-за того, что она старше всех.

#### ROMI

```python
first_order_source = pd.merge(first_order, first_source, on='uid')
first_order_source=first_order_source.reset_index()
# добавление источника к дате первого заказа
```

```python
ltv = first_order_source.merge(orders, on = 'uid')
ltv['lifetime'] = ((ltv['buy_ts'] - ltv['first_order']) / np.timedelta64(1, 'M')).round().astype(int)
orders_lifetime = ltv.pivot_table(index = 'source_id', columns = 'lifetime', values = 'revenue', aggfunc = 'count').fillna(0)
buyers_lifetime = ltv.pivot_table(index = 'source_id', columns = 'lifetime', values = 'uid', aggfunc = 'nunique').fillna(0)
ltv = ltv.pivot_table(index = 'source_id', columns = 'lifetime', values = 'revenue', aggfunc = 'sum').fillna(0)
ltv = ltv.cumsum(axis = 1)
# вывод LTV
```

```python
cohort_sizes = first_order_source.groupby('source_id').agg({'uid': 'nunique'}).rename(columns = {'uid': 'cohort_size'})
ltv = ltv.merge(cohort_sizes, on = 'source_id')
ltv = ltv.div(ltv['cohort_size'], axis = 0).drop(columns = 'cohort_size')
# вывод размеров когорт
```

```python
cac = costs.groupby('source_id').agg({'costs': 'sum'}).reset_index().merge(cohort_sizes.reset_index(), on = 'source_id')
cac['cac'] = cac['costs'] / cac['cohort_size']
cac = cac.set_index('source_id')
# вывод САС
```

```python
romi = ltv.merge(cac[['cac']], on = 'source_id')
romi = romi.div(romi['cac'], axis = 0).drop(columns = 'cac')
# вывод ROMI
```

```python
plt.figure(figsize=(16, 7))
plt.title('ROMI')
sns.heatmap(romi, annot=True, fmt='.2%', linewidths=1, cmap="YlGnBu")
plt.ylabel('Источника трафика')
plt.xlabel('Прошло месяцев с первого заказа')
plt.show()
# тепловая карта ROMI
```

На тепловой карте выделяется 3 источник трафика с худшей окупаемостью, за год он вернул только 35.6% затрат. Первый канал наоборот почти полностью отбил вложения на второй месяц жизни. 2, 5 и 9 источники также показали себя неплохо окупившись в среднем за полгода. 4 и 10 источники не успели окупиться за исследуемый период.

## Выводы и рекомендации

### Выбор источников трафика

Стоит перераспределить бюджет с 3 источника трафика, т.к. CAC по нему самый большой (13.5 y.e.), как и процент от всего бюджета на маркетинг (больше 42%) и вложеные в него средства за год не отбились и на 40%. В первую очередь бюджет стоит вложить в 1 источник, с него приходит самая конверсионная аудитория. Второй канал тоже возвращает инвестиции довольно быстро (примерно за 6 месяцев), несмотря на высокую стоимость привлечения клиента(12.2 y.e.) Также стоит попробовать больше вложить в 5 и 9 источники, привлечение клиентов с них обходится не так дорого (7.5 и 5 y.e. соответственно) и окупается в среднем за 8 месяцев. Также интерес представляет 4 источник, если получится увеличить конверсию по нему, с него пришло довольно много покупателей с небольшим САС.

### Выводы по метрикам

Сервисом пользуются редко, покупки в основном совершаются при первом посещении, причем в течении 1 минуты после захода. Средний чек 1 покупки - 5 y.e. В среднем чтобы экономика сходилась по клиентам уже через полгода, необходима стоимость привлечения одного клиента на уровне - 7-8 y.e.. Стоит подумать над увеличением количества покупок одним пользователем, т.к. сервис располагает к этому, а за год на одного покупателя приходится всего 1.3 заказа. Есть смысл увеличить рекламный бюджет на декабрь, потому что спрос в этом месяце сильно растет. В апреле есть заметный спад количества сессий, возможно это связано с сокращением затрат на рекламу, но стоит уточнить, не было ли проблем с сайтом в это время.

### Выводы по когортному анализу

Наибольший интерес вызывает когорта 2017-09, покупатели из нее приносят больше всего денег, видно, что данная когорта в декабре сильно вырвалась вперед по выручке, причём подобных увеличений со стороны других когорт не наблюдалось. Начиная с сентября бюджеты на маркетинг увеличились, возможно, поведение когорты является следствием особенностей рекламных кампаний запущенных в сентябре. Стоит уточнить, не были ли предоставлены особые условия для новых пользователей в этом месяце. Также могло сработать попадание в необходимую целевую аудиторию в этом месяце.
